As we can see the result is a list not easy to manipulate for computation so we willtransform it into a data frame:

```python
import pandas as pd
for recipe in e.search_recipe("onion and chicken and tomoatoes"):   #go through the list of all recipes
    Nutrients=recipe.totalNutrients
    NutrientsList =[]
    recipe_name = recipe
    for n in Nutrients:                                             #for each recipe go through all nutrients
        NutrientsList+=[str(n)]                                     #transforming Nutrient type into a string
    df = pd.DataFrame(data = NutrientsList, columns=["nutrient_list"])      #creating df from the list of nutrients
    df[["Nutrient_notclean","tmp"]]=df.nutrient_list.str.split("*",expand=True)      #splitting name and measurement using '*' also to avoid issues with Vitamins
    df[["Value","Unit"]]=df.tmp.str.split(expand=True)                   #splitting measurement into value and unit
    Nutrients2 =[]                                                       #creating a new list to remove space at the end of the nutrients
    for n in df.Nutrient_notclean:
        Nutrients2+=[n.strip()]
    df["Nutrient"]=Nutrients2                                            #adding this column to df
    df["recipe"]=recipe_name                                             #adding column with recipe name
    df["Value"]=df["Value"].astype(float)                                #converting Value fro string to flaot to be able to do pivot
```

```python
 df.pivot_table('Value', 'recipe', 'Nutrient')
```

```python
dftot=pd.DataFrame(columns=["nutrient_list","Nutrient_notclean","tmp","Value","Unit","Nutrient","recipe"])
for recipe in e.search_recipe("onion and chicken"):   #go through the list of all recipes
    Nutrients=recipe.totalNutrients
    NutrientsList =[]
    recipe_name = recipe
    for n in Nutrients:                                             #for each recipe go through all nutrients
        NutrientsList+=[str(n)]                                     #transforming Nutrient type into a string
    df2 = pd.DataFrame(data = NutrientsList, columns=["nutrient_list"])      #creating df from the list of nutrients
    df2[["Nutrient_notclean","tmp"]]=df2.nutrient_list.str.split("*",expand=True)      #splitting name and measurement using '*' also to avoid issues with Vitamins
    df2[["Value","Unit"]]=df2.tmp.str.split(expand=True)                   #splitting measurement into value and unit
    Nutrients2 =[]                                                       #creating a new list to remove space at the end of the nutrients
    for n in df2.Nutrient_notclean:
        Nutrients2+=[n.strip()]
    df2["Nutrient"]=Nutrients2                                            #adding this column to df
    df2["recipe"]=str(recipe_name)                                            #adding column with recipe name
    df2["Value"]=df2["Value"].astype(float)                                #converting Value for string to flaot to be able to do pivot
    dftot=dftot.append(df2)
```
